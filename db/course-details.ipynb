```python
import pandas as pd 
import numpy as np
from faker import Faker
import datetime
import random
import langdetect

fake = Faker('en_IN')
```

```python
df = pd.read_csv('../Exampleset/sample_course.csv')
df.head()
```

```python
df.count()
```

```python
df = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0','url','avg_rating','avg_rating_recent','is_recently_published', \
    'published_at','last_update_date', 'published_title', 'objectives_summary', 'created_at', 'is_paid'])
```

```python
df.columns = ['course_id','title','instructors','url','headline','num_subscribers', 'rating','num_reviews', \
    'total_lectures','practice_tests','closed_caption','course_level','content_duration','category']
```

```python
df.head()
```

```python
len(df)
```

```python
published_dates = [fake.date_time_between_dates(datetime.date(2020,1,1), datetime.date(2021,12,31)).strftime('%Y-%m-%d %H:%M:%S') for i in range(len(df))]
```

```python
# at published date
df['published_date'] = published_dates
```

```python
def updated_at_fun(row):
    new_date = fake.date_time_between_dates(datetime.date(2022,1,1), datetime.date(2024,1,15)) #.strftime('%Y-%m-%d %H:%M:%S')
    old_date = datetime.datetime.strptime(row, '%Y-%m-%d %H:%M:%S')
    difference_in_days = new_date - old_date
    if difference_in_days.days > 45:
        return new_date.strftime('%Y-%m-%d %H:%M:%S')
    else:
        updated_at_fun(row)
    
```

```python
df['last_updated'] = df['published_date'].apply(updated_at_fun)
```

```python
df['url'] = df['url'].apply(lambda x: x.replace('udemycdn','edtech').replace('udemy','edtech'))
```

```python
df['price'] = np.random.randint(low=100, high=300, size=len(df))
```

```python
# options = ['A', 'B', 'C']
# df['New Column'] = random.choices(options, k=len(df))

df['currency'] = random.choice(['USD','INR'])
```

```python
df['course_price'] = "$ " + df['price'].astype('str') + " " + df['currency']
```

```python
df['course_language'] = df['title'].apply(lambda x: langdetect.detect(x))
```

```python
list_s = ['course_id','title','headline','category','course_level','published_date','last_updated','course_price','course_language','total_lectures','content_duration',\
    'closed_caption','practice_tests','instructors','num_subscribers','num_reviews','rating','url']

for i in df.columns:
    if i not in list_s:
        print(i)
```

```python
df = df[list_s]
```

```python
df.course_language.value_counts()
```

```python
def translated_or_not(row):
    if row != 'en':
        return random.choice(['Yes','No','YEs','YES',True,'Ys','NO',False])
    else:
        return random.choice(['Yes','No','YEs','YES',True, 'NO',False])

df['is_translated'] = df['closed_caption'].apply(translated_or_not)
```

```python
df.head()
```

```python
def translated_language(row):
    if row['is_translated'] in ['Yes','YEs','YES',True,'Ys']:
        if row['course_language']=='en':
            return list(set(random.choices(['es','hi','ru','fr','zh','ja','de','pt',],k=3)))
        else:
            return ['en','hi']
    else:
        return ''
        
    


df['translated_language'] = df.apply(translated_language, axis=1)
```

```python
df
```

```python
df = df[df['num_subscribers']>0]
```

```python
len(df)
```

```python
df.to_csv('./data/course.csv',index=False)
```
